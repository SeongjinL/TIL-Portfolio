```python
col = ["id","롤","오버워치","파판14","로스트아크","트릭스터","카트라이더","메이플스토리",
       "레식","GTA5","크레이지아케이드","그랜드체이스","하스스톤","히오스","알투비트","서든어택","마비노기","엘소드"]

df_game = pd.DataFrame(columns = col)
df_game
```

```python
#게임 전처리
for game in game_db:
    list_ = ['o', 'O','x', 'X', 'ㄴ', 'ㅇ', '❌','⭕', 'Y', 'N', 'y', 'n', '×', '○']
    temp = {}
    game['text'] = game['text'].replace(' ','')
    game['text'] = game['text'].replace('\n','')
    game['username'] = game['username'].replace(' ','')
    for i in range(len(col)):
        if i == 0:
            temp[col[i]] = game['username']
            
        elif i != 0:
            data = re.search(col[i], game['text'])
            if data :
                text = game['text'][data.span()[1]+1:data.span()[1]+2]
                if text in list_:
                    if text == 'o' or text == 'O' or text == 'ㅇ' or text == '⭕' or text == 'Y' or text == 'y' or text == '○':
                        text = 1
                        temp[col[i]] = text
                    elif text == 'x' or text == 'X' or text == 'ㄴ'or text == '❌' or text == 'N' or text == 'n' or text == '×' :
                        text = 0
                        temp[col[i]] = text
                else:
                    pass
    df_game = df_game.append(temp, ignore_index=True)
```

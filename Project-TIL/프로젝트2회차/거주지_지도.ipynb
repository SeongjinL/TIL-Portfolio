```python
import pandas as pd
import geopandas as gpd
import matplotlib.pyplot as mlp
from collections import Counter
geo_file = "./data/TL_SCCO_CTPRVN.shp"
geo = gpd.read_file(geo_file)
```

```python
live = pd.read_csv('./output/liveplace.csv')
live
```

```python
c = Counter(live['liveplace'])
c
```

```python
geo 
```

```python
c = sorted(c.items())
```

```python
for i in range(0,len(c)) :
    geo.iloc[i,2] = c[i][0]
```

```python
new_df = pd.DataFrame(list(c),
                   columns=['CTP_KOR_NM', 'COUNT'])
```

```python
aaasss =  pd.merge(geo,new_df)
aaasss
```

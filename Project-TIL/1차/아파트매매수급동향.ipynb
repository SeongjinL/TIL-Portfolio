# 아파트 매매수급지수란?

- 회원 중개업소 설문과 인터넷 매물 건수 등을 분석해

- 수요와 공급 비중을 지수화한 것

 

### 아파트 매매수급지수 해석은 어떻게?

- 0에 가까울수록 공급>수요, 

- 200에 가까울수록 공급<수요

- 100에 가까우면 수요와 공급 비중이 비슷하다는 것을 뜻함
